```python
apartment_table.loc[apartment_table['Residential complex'] == "Anar", 'District'] = "Есильский район"
apartment_table.loc[apartment_table['Residential complex'] == "Garden View", 'District'] = "Есильский район"
apartment_table.loc[apartment_table['Residential complex'] == "Самрук Towers", 'District'] = "Нура район"
apartment_table.loc[apartment_table['Residential complex'] == "GreenLine.Headliner", 'District'] = "Есильский район"
apartment_table.loc[apartment_table['Residential complex'] == "GreenLine.Flora", 'District'] = "Есильский район"
apartment_table.loc[apartment_table['Residential complex'] == "Izet", 'District'] = "Есильский район"
apartment_table.loc[apartment_table['Residential complex'] == "Besterek", 'District'] = "Есильский район"
apartment_table.loc[apartment_table['Residential complex'] == "MoD. Standard", 'District'] = "Есильский район"
apartment_table.loc[apartment_table['Residential complex'] == "MoD. Comfort", 'District'] = "Есильский район"
apartment_table.loc[apartment_table['Residential complex'] == "Aisar", 'District'] = "Есильский район"
apartment_table.loc[apartment_table['Residential complex'] == "Capital Park. Emotions", 'District'] = "Есильский район"
apartment_table
```

```python
#remove rows where Residential complex is NaN
apartment_table = apartment_table[~apartment_table['Residential complex'].isna()]
print(apartment_table['Residential complex'].isna().sum())  
```

```python
apartment_table.loc[apartment_table['Material'].isna(), 'Material'] = 'Unknown'
apartment_table
```

```python
#сalculate the most common 'Housing class' for each 'Rooms' group
modes = apartment_table.groupby('Rooms')['Housing class'].transform(lambda x: x.mode()[0] if not x.mode().empty else None)
#fill missing 'Housing class' values with the mode of their corresponding 'Rooms'
apartment_table.loc[apartment_table['Housing class'].isna(), 'Housing class'] = modes

apartment_table.info()
```

```python
apartment_table.to_csv('apartment_cleaned.csv')
```
